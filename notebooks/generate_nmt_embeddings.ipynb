```python

collect = False
collect_folder = "./results/nmt_embeddings/%s/evaluations/" % model_name

if not collect:
    embeddings = torch.load("./results/nmt_embeddings/%s/evaluations/0_238/train/embeddings.pt" % model_name)
    embeddings.shape
else:
    collected = {}
    for root, dirs, files in os.walk(collect_folder):
        for file in files:
            if file.endswith(".pt"):
                pt_path = os.path.join(root, file)
                rel_path = os.path.relpath(pt_path, collect_folder)
                rel_path = rel_path.split("/")[-1]
                try:
                    tensor = torch.load(pt_path)
                except Exception as e:
                    print(f"Could not load {pt_path}: {e}")
                    continue
                if rel_path not in collected:
                    collected[rel_path] = []
                collected[rel_path].append(tensor)
    
    aggregated = {}
    for k, v in collected.items():
        if len(v) == 1:
            aggregated[k] = v[0]
        elif len(v) > 1 and isinstance(v[0], torch.Tensor):
            try:
                aggregated[k] = torch.cat(v)
            except Exception:
                aggregated[k] = torch.stack(v)
        else:
            aggregated[k] = v
    
    sorted_indices = torch.tensor(np.argsort(aggregated["indices.pt"].numpy()))

    embeddings = aggregated["embeddings.pt"][sorted_indices]
    correct_indices = aggregated["indices.pt"][sorted_indices]
    correct_y_values = aggregated["y_value.pt"][sorted_indices]

```

```python
si = np.where(nmt_df.columns == "1")[0][0]
ei = np.where(nmt_df.columns == "272")[0][0]+1

wt_ref = nmt_df.iloc[0,si:ei]

mutations_tensor =\
[torch.tensor([int(x1 != x2) for x1, x2 in zip(wt_ref, nmt_df.iloc[i,si:ei])], dtype=torch.int) for i in range(1, nmt_df.shape[0])]

mutations_matrix = torch.stack(mutations_tensor, dim=0)
mutations_matrix.shape

no_wt = embeddings[1:,:]

```

```python
# Sanity
assert(sum(nmt_df.iloc[1:,:]["num_muts"].to_numpy() == mutations_matrix.sum(dim=1).numpy()) == nmt_df.shape[0] - 1)
```

```python
mutated_positions_embedding = embeddings[:,torch.where(mutations_matrix.sum(dim=0) != 0)[0],:]
```

```python
os.makedirs("data/nmt/embeddings/%s" % model_name, exist_ok=True)

torch.save(mutated_positions_embedding, "data/nmt/embeddings/%s/embeddings.pt" % model_name)

if not collect:
    indices = torch.load("./results/nmt_embeddings/%s/evaluations/0_238/train/indices.pt" % model_name)
    y_values = torch.load("./results/nmt_embeddings/%s/evaluations/0_238/train/y_value.pt" % model_name)
else:
    indices = correct_indices
    y_values = correct_y_values

torch.save(indices, "data/nmt/embeddings/%s/indices.pt" % model_name)
torch.save(y_values, "data/nmt/embeddings/%s/y_values.pt" % model_name)
```
